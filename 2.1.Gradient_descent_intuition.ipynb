```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
import matplotlib.pyplot as plt
plt.scatter(x, y)
```

# Apply OLS

```python
from sklearn.linear_model import LinearRegression
```

```python
l_r = LinearRegression()
l_r.fit(x, y)
```

```python
print(l_r.intercept_)
print(l_r.coef_)
```

```python
plt.scatter(x, y)
plt.plot(x, l_r.predict(x))
```

# Using Gradient Descent

Now we will apply gradient descent assuming that the slope is know i.e. m = 78.35, initial b = 0, and a learning rate of 0.01

```python
m, b = 78.35, 0
lr = 0.1
y_pred = (m*x+b).reshape(4)
```

```python
plt.scatter(x, y)
plt.plot(x, l_r.predict(x))
plt.plot(x, y_pred, color='red')
```

```python
# Iteration 1
slope = -2*np.sum(y-m*x.ravel()-b)
print(slope)
step_size = lr*slope
print(step_size)
b = b-step_size
print(b)
```

```python
y_pred_1 = (m*x+b).reshape(4)
plt.scatter(x, y)
plt.plot(x, l_r.predict(x), color='blue', label='OLS')
plt.plot(x, y_pred, color='red', label= 'b=0')
plt.plot(x, y_pred_1, color='green', label= 'b={:.2f}'.format(b))
plt.legend()
```

```python
# Iteration 2
slope = -2*np.sum(y-m*x.ravel()-b)
print(slope)
step_size = lr*slope
print(step_size)
b = b-step_size
print(b)
```

```python
y_pred_2 = (m*x+b).reshape(4)
plt.scatter(x, y)
plt.plot(x, l_r.predict(x), color='blue', label='OLS')
plt.plot(x, y_pred, color='red', label= 'b=0')
plt.plot(x, y_pred_1, color='green', label= 'b=20.93')
plt.plot(x, y_pred_2, color='purple', label= 'b={:.2f}'.format(b))
plt.legend()
```

```python
# Iteration 3
slope = -2*np.sum(y-m*x.ravel()-b)
print(slope)
step_size = lr*slope
print(step_size)
b = b-step_size
print(b)
```

```python
y_pred_3 = (m*x+b).reshape(4)
plt.scatter(x, y)
plt.plot(x, l_r.predict(x), color='blue', label='OLS (b=26.15)')
plt.plot(x, y_pred, color='red', label= 'b=0')
plt.plot(x, y_pred_1, color='green', label= 'b=20.93')
plt.plot(x, y_pred_2, color='purple', label= 'b=25.11')
plt.plot(x, y_pred_3, color='yellow', label= 'b={:.2f}'.format(b))
plt.legend()
```

```python
# Using for loop
m, b = 78.35, 0
lr = 0.1
epochs = 10
for i in range(epochs):
    slope = -2*np.sum(y-m*x.ravel()-b)
    step_size = lr*slope
    b = b-step_size
    print(b)
    y_pred = (m*x+b).reshape(4)
    plt.plot(x, y_pred, color='red', label= 'b=0')
plt.scatter(x, y)
```
