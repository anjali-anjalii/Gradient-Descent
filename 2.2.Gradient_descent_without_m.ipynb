```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1, noise=20 )
plt.scatter(x, y)
```

# Creating our own GD Regression class

```python
from sklearn.linear_model import LinearRegression
```

```python
l_r = LinearRegression()
l_r.fit(x, y)
print(l_r.coef_)
print(l_r.intercept_)
```

```python
class GDRegressor:
    def __init__(self, lr, epochs):
        self.m = 71.44
        self.b = 0
        self.lr = lr
        self.epochs = epochs
    def fit(self, x, y):
        for i in range(self.epochs):
            slope = -2*np.sum(y-self.m*x.ravel()-self.b)
            step_size = self.lr*slope
            self.b = self.b-step_size
        print(self.b)
        y_pred = (self.m*x+self.b).reshape(100)
        plt.scatter(x, y)
        plt.plot(x, y_pred, color='red', label= 'b={%.2f}.format(b)')
```

```python
gd = GDRegressor(0.01, 100)
```

```python
gd.fit(x, y)
```
